```python
### One Word, Diachronic



dftt = pd.DataFrame()

for i in list(range(1815,1906,10)):
    mfn = [m for m in list_mod if int(m[0:4]) == i][0]
    model = KeyedVectors.load_word2vec_format(mfn, binary = True)
    dft = pd.DataFrame()
    for ms in list(dict(model.most_similar('binnenland', topn=50)).keys()):

        first_list = list(dict(model.most_similar(ms, topn=50)).keys())

        first_df = pd.DataFrame()

        for term in first_list:
            tmp_df = pd.DataFrame([[key,value] for key,value in dict(model.most_similar(term, topn=50)).items()], columns = ['target', 'weight'])
            tmp_df['source'] = term
            first_df = first_df.append(tmp_df)

        dft = dft.append(first_df)
    dft['clust_year'] = i
    degree_ref = pd.DataFrame([dft['target'].value_counts()]).T

    
    odf = pd.DataFrame()

    for x in range(0,len(dft)-1):
        source = dft.iloc[x,2]
        weight = dft.iloc[x,1]
        target = dft.iloc[x,0]
        oc = int(degree_ref[degree_ref.index == target]['target'][0])

        tmp = pd.DataFrame([source, target, weight, oc]).T
        tmp.columns = ['source', 'target', 'weight', 'degree']
        odf = odf.append(tmp)
    
    
    odf['clust_year'] = i
    dftt = dftt.append(odf)
```

```python
### Multiple Words, Synchronic

year = 1865


dftt = pd.DataFrame()

mfn = [m for m in list_mod if int(m[0:4]) == year][0]
model = KeyedVectors.load_word2vec_format(mfn, binary = True)

for word in ['binnenland', 'buitenland', 'vaderland', 'europa', 'kolonie']:
    dft = pd.DataFrame()
    for ms in list(dict(model.most_similar(word, topn=50)).keys()):

        first_list = list(dict(model.most_similar(ms, topn=50)).keys())

        first_df = pd.DataFrame()

        for term in first_list:
            tmp_df = pd.DataFrame([[key,value] for key,value in dict(model.most_similar(term, topn=20)).items()], columns = ['target', 'weight'])
            tmp_df['source'] = term
            first_df = first_df.append(tmp_df)

        dft = dft.append(first_df)
    
    dftt = dftt.append(odf)
```

```python
degree_ref = pd.DataFrame([dftt['target'].value_counts()]).T
degree_ref = degree_ref[degree_ref.target > 200]
degree_ref = list(degree_ref.index)
degree_ref = [str(w) for w in degree_ref]

```

```python
dftta = dftt[dftt.target.isin(degree_ref)]
```
