```python
df = list()

for i in list(range(1845,1906,10)):
    mfn = [m for m in list_mod if int(m[0:4]) == i]
    m = KeyedVectors.load(mfn[0], mmap='r')

    print('')
    print(i)
    for w in [k for k,v in dict(m.wv.vocab).items() if "buitenlandsche" in k]:
        tmp = dict(m.most_similar(w, topn=500000))

        for x in [x for x,y in tmp.items() if y > 0.4]:
            if "internationale_" in x:
                weight = m.similarity(w,x)
                t = [w,x,weight,i]
                df.append(t)
                print(w, "  |  ", x)
```

```python
d = pd.DataFrame([[a for a,b,c,d in df], [b for a,b,c,d in df], [c for a,b,c,d in df], [d for a,b,c,d in df]]).T
d.columns = ['source', 'target', ' weight', 'year']
d['degree'] = 1
d
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def PlotEnt(year):
    dt = d[d['year'] == year]

    B = nx.Graph()

    B.add_nodes_from(list(set(dt['source'])), bipartite=0)
    B.add_nodes_from(list(set(dt['target'])), bipartite=1)

    dt = dt.reset_index(drop=True)
    list_edges = []
    for c,i in enumerate(dt['source']):
        s = i
        t = dt['target'][c]
        list_edges.append((s,t))

    B.add_edges_from(list_edges)

    # Separate by group
    l, r = nx.bipartite.sets(B)
    pos = {}

    # Update position for node from each group
    pos.update((node, (1, index)) for index, node in enumerate(l))
    pos.update((node, (2, index)) for index, node in enumerate(r))

    plt.figure(3,figsize=(6,12)) 

    #colors = range(len(list_edges))


    nx.draw(B, 
            with_labels = True, 
            node_size = 10, 
            font_size = 8,
            edge_color = "lightsteelblue", 
            #edge_cmap = plt.get_cmap('Blues'),
            node_color = "lightsteelblue",
            pos = pos)
    
    plt.show()
    plt.tight_layout()
```

```python
PlotEnt(1845)
```
