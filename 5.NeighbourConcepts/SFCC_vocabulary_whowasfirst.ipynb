```python
df1 = pd.read_csv("bigrams_internationale.csv")
df2 = pd.read_csv("bigrams_buitenlandsche.csv")
```

```python
list_birth = list()

for ngram in df1.ngram:
    first_year = df1[df1.ngram == ngram]
    first_year = first_year.loc[:, (first_year != 0).any(axis=0)]
    first_year = int(first_year.columns[1])
    
    first_year_df1 = first_year
    
    if ngram in list(df2.ngram):
        first_year = df2[df2.ngram == ngram]
        first_year = first_year.loc[:, (first_year != 0).any(axis=0)]
        first_year = int(first_year.columns[1])
    
        first_year_df2 = first_year
    
    if first_year_df1 < first_year_df2 and first_year_df2 - first_year_df1 <= 10:
        print(ngram + " | buitenlandsche: " + str(first_year_df1) + " | nationale: " + str(first_year_df2))
        list_birth.append(first_year_df1)
    
```
