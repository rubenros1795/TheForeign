```python
import os, re, glob, pandas as pd, math, random, nltk
from nltk.collocations import *
from tqdm import tqdm
```

```python
os.chdir("D://Scriptie//Data//lines//cleaned")
list_txt = glob.glob('*.txt')
list_txt = [f for f in list_txt if f[5:12] == "cleaned"]
```

```python
def file_len(fname):
    with open(fname, encoding = 'utf-8') as f:
        for i, l in enumerate(f):
            pass
    return i + 1
```

```python
def ImportTokensYear(year):
    fn = [f for f in list_txt if int(f[0:4]) == year][0]
    
    filesize = int(os.stat(fn).st_size) / 1000000
    filelen = file_len(fn)
    
    
    if filesize > 80:
        samplesize = round(80 * filelen / filesize)
        with open(fn, encoding = 'utf-8') as f:
            lines_year = random.sample(f.read().split('\n'),samplesize)    
    
    else:
        with open(fn, encoding = 'utf-8') as f:
            lines_year = f.read().split('\n')
    
    lines = [line.split(' ') for line in lines_year]
    lines = [item for sublist in lines for item in sublist]
    return lines
```

```python
def TokensToPMI(year):

    bigram_measures = nltk.collocations.BigramAssocMeasures()
    tokens = ImportTokensYear(year)
    finder = BigramCollocationFinder.from_words(tokens,12)
    finder.apply_freq_filter(5)
    dfr = pd.DataFrame()
    pmi = dict(finder.score_ngrams(bigram_measures.pmi))
    
    #for key, value in pmi.items():
        
        #if key[0] in list_goal or key[1] in list_goal or key[0] in list_results or key[1] in list_results:
        #tmp = pd.DataFrame([list(key)], columns = ['w1', 'w2'])
        #tmp['pmi'] = value
        #dfr = dfr.append(tmp)
        #else:
         #   continue

    #dfr['year'] = year
    return pmi
```

```python
dft = pd.DataFrame()

for year in tqdm(list(range(1815,1830))):
    tmp = TokensToPMI('buitenland', year)
    dft = dft.append(tmp)
```
