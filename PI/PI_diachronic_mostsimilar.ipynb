```python
df = list()

for i in list(range(1815,1906,10)):
    mfn = [m for m in list_mod if int(m[0:4]) == i]
    m = KeyedVectors.load(mfn[0], mmap='r')

    print('')
    print(i)
    for w in [k for k,v in dict(m.wv.vocab).items() if "internationale_" in k]:
        tmp = dict(m.most_similar(w, topn=1000))

        for x in [x for x,y in tmp.items() if y > 0.4]:
            if "buitenlandsche_" in x:
                t = [w,x,i]
                df.append(t)
                print(w, "  |  ", x)
```

```python
d = pd.DataFrame([[a for a,b,c in df], [b for a,b,c in df], [c for a,b,c in df]]).T
```

```python
d.columns = ['source', 'target', 'year']
```
