** Data Preparation **

```python
df = pd.read_csv('creativity_buitenlandsche.csv', sep = " ")
```

```python
df['count'] =df['count'].astype(float)
df['years'] =df['years'].astype(float)
```

** Scatterplot**

```python
x_values = df['years']
y_values = df['count']
x_lab = "Years"
y_lab = "Absolute Frequency"
title = "Creativity Scores for 'Buitenlandsche'-vocabulary"
color_point = "grey"
rm = y_values.rolling(window=5).mean()

mpl_fig = plt.scatter(x = x_values, y = y_values, c = color_point, s = 10)
plt.plot(x_values, rm)
plt.title(title)
#plt.savefig("productivity-buitenlandsche.png", format = 'png', dpi = 800)
plt.show()
```

** Stacked Area Plot (for clusters) **

```python
os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//Visualizations//CH1//source-files")
```

```python
col = sns.color_palette("tab20b", 11)
```

```python
for i in [6,8,10]:
    fn = "buitenlandsche_clusters_" + str(i) + ".csv"
    df = pd.read_csv(fn)
    
    x_lab = "Years"
    y_lab = "Absolute Frequency"
    title = "Cluster Distribution of 'Buitenlandsche'-vocabulary with " + str(i) + " Clusters"
    
    
    fig, ax = plt.subplots(facecolor='white',figsize=(24, 8))
    labels = list(df.columns[:-2])
    plt.stackplot(list(df.year), *[df[col] for col in labels], baseline="wiggle", colors = col)
    ax.set_facecolor("white")
    box = ax.get_position()
    #hfont = {'fontname':'Helvetica'}

    ax.set_position([box.x0, box.y0, box.width * 0.8, box.height])
    lgd = ax.legend(labels, loc='upper center', bbox_to_anchor=(0.5, -0.06))
    
    for ticks in ax.xaxis.get_ticklines() + ax.yaxis.get_ticklines():
        ticks.set_color("white")
    for pos in ['top', 'bottom', 'right', 'left']:
        ax.spines[pos].set_edgecolor("white")
    
    plt.yticks(fontsize=14)
    plt.xticks(fontsize=18)

    plt.title(title, fontsize=22)


    #plt.title('title',**hfont)
    #plt.xlabel('xlabel', **hfont)
    
    fn = "cluster-plot-" + str(i) + ".png"
    os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//Visualizations//CH1")
    #fig.savefig(fn, facecolor=fig.get_facecolor(), dpi=800,bbox_extra_artists=(lgd,), bbox_inches='tight')
    os.chdir("C://Users//Ruben//Documents//GitHub//TheForeign//Visualizations//CH1//source-files")
```
