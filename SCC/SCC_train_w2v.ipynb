```python
from nltk.tokenize import sent_tokenize
from gensim.models import Word2Vec
import nltk
from nltk.tokenize import sent_tokenize
from nltk.corpus import stopwords
import glob
import os
from os import path
import string
import numpy
import re
from __future__ import absolute_import, division, print_function
import pandas as pd
import multiprocessing
import gensim
```

```python
os.chdir("C://Users//Ruben//Documents//Scriptie//test-environment")
files = glob.glob('*.txt')
files
```

```python
sentences = list()

for i in files:
    if os.stat(i).st_size != 0:
        with open(i) as f:
            content = f.read().splitlines()
        for sent in content:
            sentence = sent.split()
            sentences.append(sentence)

model = Word2Vec(sentences, size=150, window=10, iter = 12, min_count=3, workers=3)
model.train(sentences,total_examples=len(sentences),epochs=10)
```

```python
ms = model.most_similar("b", topn=10)

ttl = list()

for i in ms: 
    tmp_ms = model.most_similar(i[0], topn=10)
    for x in tmp_ms:
        tmp_ms_2 = model.most_similar(x[0], topn=10)
        for z in tmp_ms_2:
            ttl.append(z[0])
```

```python
from collections import Counter
```

```python
df = pd.DataFrame(list(Counter(ttl).keys()))
df['dist'] = list(Counter(ttl).values())
```
