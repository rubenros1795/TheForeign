```python
os.chdir("D://Scriptie//Data//lines//cleaned")
all_txt = glob.glob('*.txt')
all_txt = [f for f in all_txt if f[6:12] == 'leaned']
```

```python
stopz = [word.split() for word in open("C:\\Users\\Ruben\\Documents\\Artikelen\\Joris\\stopwords-nl.txt", 'r', encoding = "utf-8").readlines()]
stopz = [item for sublist in stopz for item in sublist]
```

```python
search_term = "binnen"

results = list()

for year in range(1815, 1914+1):
    list_lines_year = list()

    files = [s for s in all_txt if str(year) in s]

    for file in files:
        lines = [line.rstrip('\n') for line in open(file, encoding = 'utf-8')]
        lines = [line for line in lines if search_term in line]
            
            
        for line in lines:
            line_tok = line.split(" ")
            positions = [idx for idx, s in enumerate(line_tok) if search_term in s]

            for pos in positions:
                position_binn = pos
                pot_pos_buit = pos + 2
                
                if pot_pos_buit >= len(line_tok):
                    continue
                if line_tok[pot_pos_buit] == "buitenlandsche":
                    tmp = [year, " ".join(line_tok[position_binn:position_binn+4])]
                    print(" ".join(line_tok[position_binn:position_binn+5]))
                    results.append(tmp)
```

```python
df = pd.DataFrame(results, columns = ['year', 'p'])
```

```python
os.chdir("C:\\Users\\Ruben\\Documents\\Scriptie\\Data\\kwic_dfs")
df.to_csv("binnenlandsche_x_en_buitenlandsche_y.csv")
```
