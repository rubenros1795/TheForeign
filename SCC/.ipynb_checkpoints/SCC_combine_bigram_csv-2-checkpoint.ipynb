**Create Yearly .CSV per Newspaper**

The newspapers are sometimes ngramized per month or day. The script below merges them per year. 

```python
os.chdir("C://Users//Ruben//Documents//Scriptie//Data//cngwf")
```

```python
list_files = glob.glob('*.csv')
list_names = list(set([f[0:4] for f in list_files]))
list_years = list(set([int(f.split("_")[2][0:4]) for f in list_files]))
```

```python
for name in list_names:
    tmp_list_names = [f for f in list_files if name in f]
    
    for year in list(set([int(f.split("_")[2][0:4]) for f in tmp_list_names])):
        tmp_list_year_name = [x for x in tmp_list_names if int(x.split("_")[2][0:4]) == year]
        
        df_year_name = pd.DataFrame(columns = ['ngram', 'year', 'count'])
        
        for file in tmp_list_year_name:
            df = pd.read_csv(file)        
            df = df.groupby(['ngram','year']).agg('sum')
            df = df.stack().reset_index().drop('level_2', axis = 1)
            df.columns = ['ngram', 'year', 'count']
            df_year_name = df_year_name.append(df)
        
        df['year'] = df['year'].astype(str)
        df['year'] = df['year'].str.slice(0,4)
        df_year_name = df_year_name.groupby(['ngram','year']).agg('sum')
        df_year_name = df_year_name.stack().reset_index().drop('level_2', axis = 1)
        df_year_name.columns = ['ngram', 'year', 'count']
        
        os.chdir("C://Users//Ruben//Documents//Scriptie//Data//cngwf//output")
        filename = str(name) + "_" + str(year) + '.csv'
        df_year_name.to_csv(filename, index = False)
        os.chdir("C://Users//Ruben//Documents//Scriptie//Data//cngwf")
        print(filename + " written")
```

**Create CSV's pe Year**

The newspapers are sometimes ngramized per month or day. The script below merges them per year. 

```python
list_years= list(range(1815,1915))
os.chdir("C://Users//Ruben//Documents//Scriptie//Data//cngwf//output")
list_csv = glob.glob('*.csv')
```

```python
for year in list_years:
    list_csv_year = [f for f in list_csv if int(f[5:9]) == year]
    
    df_year = pd.DataFrame()
    
    for file in list_csv_year:
        df = pd.read_csv(file)        
        df['year'] = df['year'].astype(str)
        df['year'] = df['year'].str.slice(0,4)
        df['year'] = df['year'].astype(int)
        df = df.groupby(['ngram','year']).agg('sum')
        df = df.stack().reset_index().drop('level_2', axis = 1)
        df.columns = ['ngram', 'year', 'count']
        df_year = df_year.append(df)
    
    df_year = df_year.groupby(['ngram','year']).agg('sum')
    df_year = df_year.stack().reset_index().drop('level_2', axis = 1)
    df_year.columns = ['ngram', 'year', 'count']
    
    os.chdir("C://Users//Ruben//Documents//Scriptie//Data//cngwf//output//years")
    filename = str(year) + "_unigrams.csv"
    df_year.to_csv(filename, index = False)
    os.chdir("C://Users//Ruben//Documents//Scriptie//Data//cngwf//output")
```
