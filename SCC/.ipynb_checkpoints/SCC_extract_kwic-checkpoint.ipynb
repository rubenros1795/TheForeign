```python
import csv
import os
import glob
import pandas as pd
import nltk
import string
import re
from nltk import ngrams
import multiprocessing
import time
```

```python
os.chdir("C://Users//Ruben//Documents//Scriptie//Data//test-environment")
list_txt = glob.glob("*.txt")
```

```python
list_years = ' '.join(str(e) for e in list(range(1815, 1915))).split()
```

```python
kwics_year = dict()

for year in list_years:
    files = [s for s in list_txt if year in s]
    
    if len(files) == 0:
        continue
            
    for file in files:
        lines = [line.rstrip('\n') for line in open(file, encoding = 'utf-8')]
        lines = [line for line in lines if "buitenland" in line]
        
        for i in range(0, len(lines) - 1):
            
            line_tok = lines[i].split(" ")
            positions = [idx for idx, s in enumerate(line_tok) if 'buitenland' in s]
            
            for pos in positions:
                p = int(pos)
                s = p - 24
                e = p + 24
                
                if s < 0:
                    s = 0
                if e > (len(line_tok) - 1):
                    e = (len(line_tok) - 1)
            
                selection = " ".join(line_tok[s:e])
                key = file[0:4] + "_" + year + "_" + str(i) + "_" + str(pos)
                tmp = {key:selection}
                kwics_year.update(tmp)
        
   
```

```python
kwic_df = pd.DataFrame(list(kwics_year.keys()))
kwic_df['line'] = pd.DataFrame(list(kwics_year.values()))
kwic_df.columns = ['id', 'kwic']
kwic_df
```
